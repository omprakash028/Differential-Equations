```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy.integrate import odeint, solve_ivp
```

```python
alpha = 3;
beta = 5;
```

```python
def dvdt(t,v):
    return alpha*v**2 - beta
v0 = 0
```

```python
t = np.linspace(0,1,100)
sol_m1 = odeint(dvdt, y0 = v0, t = t, tfirst = True)
sol_m2 = solve_ivp(dvdt, t_span = (0, max(t)), y0 = [v0], t_eval = t)
```

```python
sol_m1.T
```

```python
sol_m2.y
```

```python
v_sol_m1 = sol_m1.T[0]
v_sol_m2 = sol_m2.y[0]
```

```python
plt.plot(t, v_sol_m1, label = 'odeint')
plt.plot(t, v_sol_m2, label = 'solve_ivp')
#plt.xlabel("t", fontname = "Times New Roman", fontsize = 12)
#plt.ylabel("v(t)", fontname = "Times New Roman", fontsize = 12)
plt.title('First ODEs', fontsize = 12)
plt.xlabel('t', fontsize = 12)
plt.ylabel('v(t)', fontsize = 12)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.legend(fontsize = 12)
plt.grid()
plt.show()
```
