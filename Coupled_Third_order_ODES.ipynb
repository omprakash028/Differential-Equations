# **Solving Two Coupled Third-Order Differential Equations**

The system of coupled third-order differential equations is given by


$$\begin{equation}
\begin{split}
 \dddot{x}_2 &= -\left(\ddot{x}_1\right)^3 + \dot{x}_2 + x_1 + \sin(t) \\
\dddot{x}_1 &= -2\left(\dot{x}_2\right)^2 + x_2
\end{split}
\end{equation}$$

**Define the State Variables**

$$\begin{equation}
\begin{split}
v_1 &= \dot{x}_1 \\
a_1 &= \ddot{x}_1 = \dot{v}_1\\
v_2 &= \dot{x}_2\\
a_2 &= \ddot{x}_2 = \dot{v}_2
\end{split}
\end{equation}$$

Using these definitions, the governing equations become

$$\begin{equation}
\begin{split}
\dot{a}_2 &= -a_2^{3} + v_2 + x_1 + \sin(t)\\
\dot{a}_1 &= -2v_2^{2} + x_2
\end{split}
\end{equation}$$

So,

 **State Vector Representation**

Define the state vector as

$$
\vec{S}=
\begin{bmatrix}
x_1\\
v_1\\
a_1\\
x_2\\
v_2\\
a_2
\end{bmatrix}
$$

Therefore,

$$
\frac{d\vec{S}}{dt}
=
\begin{bmatrix}
\dot{x}_1\\
\dot{v}_1\\
\dot{a}_1\\
\dot{x}_2\\
\dot{v}_2\\
\dot{a}_2
\end{bmatrix}
=
\begin{bmatrix}
v_1\\
a_1\\
-2v_2^{2}+x_2\\
v_2\\
a_2\\
-a_1^{3}+v_2+x_1+\sin(t)
\end{bmatrix}.
$$

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy.integrate import odeint, solve_ivp
```

```python
def dsdt(t, S):
    x1, v1, a1, x2, v2, a2 = S
    return [v1,
            a1,
            - 2*v2**2 + x2,
            v2,
            a2,
            - a1**3 + v2 + x1 + np.sin(t)]
x1_0 = 0
v1_0 = 0
a1_0 = 0
x2_0 = 0
v2_0 = 0
a2_0 = 0
S0 = (x1_0, v1_0, a1_0, x2_0, v2_0, a2_0)
```

```python
t = np.linspace(0, 1, 100)
sol = odeint(dsdt, y0 = S0, t= t, tfirst=True)
```

```python
sol
```

```python
x1_sol = sol[:, 0]
v1_sol = sol[:, 1]
a1_sol = sol[:, 2]
x2_sol = sol[:, 3]
v2_sol = sol[:, 4]
a2_sol = sol[:, 5]
```

```python
plt.plot(t, x1_sol, label = 'x1')
plt.plot(t, x2_sol, label = 'x2')
plt.title('Third order ODEs solutions', fontsize = 12)
plt.xlabel('t', fontsize = 12)
plt.ylabel('v(t)', fontsize = 12)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.legend(fontsize = 12)
plt.xlabel('time (t)', fontsize = 12)
plt.ylabel('Amplitude', fontsize = 12)
plt.grid()
plt.legend()
plt.show()
```
