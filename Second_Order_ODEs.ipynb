Python does not have functions to directly solve second order ODEs. But any seconcd order ODEs can be convereted into two first order ODEs.

Consider,

$$ \ddot{x} = - \dot{x}^2 + \sin{(x)}$$

We can convert this into two first order ODEs as follows:

$Step 1 :$ Take $x$ (this is what we we're trying to solve for ). Then, define $\dot{x} = v$ so that $v$ becomes a new variable.

$Step 2 :$ Note that $\dot{x} = v$ is one differential equation.

$Step 3 :$ Since, $\dot{v} = \ddot{x}= - \dot{x}^2 + \sin{(x)} = - v^2 + \sin{(x)}$, we get another differential equation.

Oure two differential equations are:


$$\dot{x} = v \\
\dot{v} = - v^2 + \sin{(x)}$$

These two ODEs are two coupled first order equations. They require an initial conditions $(x_0, v_0)$.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy.integrate import odeint, solve_ivp
```

```python
def dsdx(x, S):
    x, v = S
    return[v,
           - v**2 + np.sin(x)]
x0 = 0
v0 = 5
S0 = (x0, v0)
```

```python
x = np.linspace(0, 1, 100)
sol = odeint(dsdx,y0 =S0, t = x, tfirst = True)
```

```python
sol
```

```python
x_sol = sol[:,0]
v_sol = sol[:,1]
```

```python
plt.plot(x, x_sol, label = 'x(x)')
plt.plot(x, v_sol, label = 'v(x)')
plt.xlabel('x')
plt.ylabel('Amplitude')
plt.title('Second order ODE')
plt.grid()
plt.legend()
plt.show()
```
